###### In dataset some columns may be un important , play no contribution in ML algorithms, some may created by merging two columns,
###### Need to select only those columns that play crucial  in ML algos,
##### Choosing the important features from dataset called feature selection


###### For feature selection domain knowledge is necessary , feature can be select by various techniques, 

#### 1-- Droping Constant Features

```python
import pandas as pd
data  = pd.DataFrame({"A" : [1,2,4,1,2,4],
                      "B" : [4,5,6,7,8,9],
                      "C" : [0,0,0,0,0,0],
                      "D" : [1,1,1,1,1,1]})
```

```python
data.head()
```

##### Variance Threshold:  Feature Selector that removes all low-Variance featutres. This algorithm only looks at features(X) , not the desired outputs(Y), and thus can be used for unsupervised learning

```python
from sklearn.feature_selection import VarianceThreshold
var_threshold = VarianceThreshold(threshold = 0) # it can b 0.2 , 0.3 , etcc
var_threshold.fit(data)
```

```python
var_threshold.get_support() # False indicates Zero variance
```

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_threshold.get_support()]]
print(len(constant_columns))
```

```python
data = data.drop(constant_columns, axis =1)
```

##### With large dataset

###### divide the  columns into dependent and independent , for exp : id , age is independent and salary is dependent, We can do this by  --> X = df[['id', 'age']], Y = df['salary']
##### its bad practice to train all the dataset , good to train 80% of test and test remainig 20%. train_test_split() has parameter 'test_size = 0.2' , means 20% dataset is for testing.
##### every time we execute this 'train_test_split()' output differ , for sample to be remain same we add parameter 'random_state = /anyvalue/' 


```python
df = pd.read_csv('Santander.csv', nrows=10000)
df.shape
```

```python
df.head(3)
```

### need to divide in independent and dependent features because variance threshold applies only on independent

```python
X = df.drop(labels = ['TARGET'],  axis =1) # expect  TARGET , all are the independent features 
Y = df['TARGET']
```

```python
Y
```

#### Test and Train
#### This splitting is only done in supervised ML

```python
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state= 0 )
```

```python
X_train.shape , X_test.shape 
```

#### Apply Variance threshold

```python
var_threshold1 = VarianceThreshold(threshold= 0 ) # columns have 0 variance
var_threshold1.fit(X_train) # fit only X_train , X_Test to avoid overfitting
```

```python
var_threshold1.get_support() # false means 0 variance
```

```python
sum(var_threshold1.get_support()) #out of 370 , 284 have true variance
```

```python
constant_columns = [column for column in X_train.columns if column not in  X_train.columns[var_threshold1.get_support()]]
print(len(constant_columns))
```

```python
X_train= X_train.drop(constant_columns, axis=1)
```
