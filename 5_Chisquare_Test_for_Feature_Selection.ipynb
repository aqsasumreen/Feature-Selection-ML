#### Need to understand the relationship with respect to the categorical variables then use Chisquare distribution

##### This score can be used to select the n_features with the highest values for the test 'Chisquare ' statistic from X, which must contain only non-negative features such as booleans or frequencies .

###### import dataset with more categorical features

```python
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df = df[['sex', 'embarked', 'alone', 'pclass', 'survived']]
df
```

##### Always before applying chisquare test we need to apply label encoding

```python
from sklearn.preprocessing import   LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df["sex"])
le
```

```python
df["sex"] = le.transform(df["sex"])
```

```python
le.fit(df["embarked"])
df["embarked"] = le.transform(df["embarked"])
```

```python
le.fit(df["alone"])
df["alone"] = le.transform(df["alone"])
```

```python
le.fit(df["pclass"])
df["pclass"] = le.transform(df["pclass"])
```

```python
le.fit(df["survived"])
df["survived"] = le.transform(df["survived"])
```

```python
df
```

```python
X = df.drop(labels = ['survived'],  axis =1) # expect  TARGET , all are the independent features 
Y = df['survived']
```

```python
X
```

#### Test train split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y , test_size= 0.3, random_state= 0)
```

```python
X_train.shape, X_test.shape
```

##### Perform chi2 tets
##### chi2 returns 2 values , Fscore and pvalue 

```python
from sklearn.feature_selection import chi2
f_p_values = chi2(X_train, Y_train)
```

```python
f_p_values
```

###### -->The more higher the value of f_score the more important the feature is
###### -->The lesser the p_value the more important the feature is

```python
import pandas as pd
p_values = pd.Series(f_p_values[1])
p_values.index = X_train.columns
p_values.sort_index(ascending= False)
p_values
```
